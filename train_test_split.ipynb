```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
quote_data = pd.read_csv('quote_data.csv')
```

```python
quote_data = quote_data.dropna()
```

```python
# Split into train and test
train, test = train_test_split(quote_data, test_size = 0.2, random_state = 42)
```

```python
train.to_csv('train_quotes.csv')
test.to_csv('test_quotes.csv')
```
